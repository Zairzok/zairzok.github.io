## Example with interactivity

``` python
(

        {

            "encoding": {
                "x":{"field":"Agency Name", "type":"ordinal"},
                "y":{"aggregate":"mean", "field":"Year Constructed", "type": "temporal"},
            }
        },
        {
            "transform": [{"filter": {"param": "brush"}}],
            "description": "Histogram Attempt #1",
            "height":"300",
            "mark": "bar",
            "encoding": {
                "x": {"bin": True, "field": "Rep Dist", "type": "quantitative", "title":"Count of Rep Dist"},
                "y": {
                    "aggregate":"count",
                    "type": "quantitative",
                    "axis": {"title": "Histogram"}
                }
            }
        }
    ]
)
```

````python
chart1 = alt.Chart(mobility_url).mark_point().encode(
    alt.X("Agency Name:O"),
    alt.Y("Year Constructed:T"),
).properties(
    height=400
)

chart1
````

````python
chart1.save('C:/Users/japet/zairzok.github.io/assets/json/chart1.json')
````

````python
chart2 = alt.Chart(mobility_url).mark_bar().encode(
    alt.X("Rep Dist:Q",bin=True,axis=alt.Axis(title="Count of Dist Rep")),
    alt.Y('count()',axis=alt.Axis(title='Histogram')),
)

chart2
````

````python
brush = alt.selection_interval(encodings=['x','y'])

chart1 = alt.Chart(mobility_url).mark_point().encode(
    alt.X("Agency Name:O"),
    alt.Y("Year Constructed:T"),
).properties(
    height=800
).add_selection(
brush
)

chart2 = alt.Chart(mobility_url).mark_bar().encode(
    alt.X("Rep Dist:Q",bin=True,axis=alt.Axis(title="Count of Dist Rep")),
    alt.Y('count()',axis=alt.Axis(title='Histogram')),
).transform_filter(
    brush
)

chart = chart1.properties(width=300) | chart2.properties(width=300)

chart
````

````python
chart.save('C:/Users/japet/zairzok.github.io/assets/json/chart_interactive.json')
````
